```python
# Перестановки: Сколько четырехзначных чисел можно получить из цифр 1, 2, 3, 4?
count = 0
for p in itertools.permutations('1234', 4):
    print(''.join(str(x) for x in p))
    count = count + 1
print("total permutations:", count)
```

```python
# Размещение: Сколько двухзначных чисел можно получить из цифр 1, 2, 3, 4?
count = 0
for a in itertools.permutations('1234', 2):
    print(''.join(str(x) for x in a))
    count = count + 1
print("total permutations:", count)
```
