###  3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,	
эллипса,
гиперболы.

```python
# окружность
from pylab import rcParams

rcParams['figure.figsize'] = 3, 3
x = []
y1 = []
y2 = []
r = 1
for i in range(-100, 101):
    x_ = i/100
    x.append(x_)
    y1.append(np.sqrt(r**2-x_**2))
    y2.append(-np.sqrt(r**2-x_**2))
    
plt.plot(x, y1, color = "red")
plt.plot(x, y2, color = "red")
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# эллипс
x = []
y1 = []
y2 = []
a = 4
b = 2
for i in range(-500, 501):
    x_ = i/100
    x.append(x_)
    y1.append(np.sqrt((1-x_**2/a**2)*b**2))
    y2.append(-np.sqrt((1-x_**2/a**2)*b**2))
    

plt.plot(x, y1, color = "green")
plt.plot(x, y2, color = "green")
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('scaled')
```

```python
# гипербола
x = []
y1 = []
y2 = []
a = 0.5
b = 0.3
for i in range(-300, 301):
    x_ = i/100
    x.append(x_)
    y1.append(np.sqrt((1+x_**2/a**2)*b**2))
    y2.append(-np.sqrt((1+x_**2/a**2)*b**2))
    
#plt.axis([-3.5, 3.5, -3.5, 3.5])
plt.plot(x, y1, color = "blue")
plt.plot(x, y2, color = "blue")
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('scaled')
```

### 5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
# трехмерный график двух параллельных плоскостей

from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)

X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5 ,0.5)
X,Y = np.meshgrid(X,Y)

Z = 3*X + 2*Y

ax.plot_wireframe(X, Y, Z)
ax.plot_surface(X, Y, Z+50)

ax.scatter(0,0,0, 'z', 50, 'red')
show()
```

```python
# трехмерный график двух любых поверхностей второго порядка
fig = figure()
ax = Axes3D(fig)

X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)

Z = X**2 - 5*Y**2
Z1 = 3*X**2 + Y**2 + 200

ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)

ax.scatter(0,0,0, 'z', 50, 'red')
show()
```
